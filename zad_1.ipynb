```
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = {
    'size': ['L', 'M', 'XL', 'S', 'XS', 'S', 'S', 'M', 'L', 'XXL', 'L', 'M', 'M', 'XS', 'S', 'XL', 'L', 'S', 'M'],
    'color': ['green', 'red', 'blue', 'black', 'white', 'yellow', 'purple', 'cyan', 'black', 'black', 'white', 'blue', 'red', 'green', 'blue', 'white', 'white', 'black', 'white'],
    'gender': ['female', 'female', 'male', 'female', 'female', 'male', 'female', 'male', 'female', 'male', 'male', 'female', 'male', 'female', 'male', 'female', 'male', 'male', 'female'],
    'price': [90.0, 15.0, 35.0, 60.0, 80.0, 45.0, 70.0, 150.0, 180.0, 30.0, 20.0, 40.0, 50.0, 60.0, 20.0, 10.0, 40.0, 150.0, 180.0],
    'weight': [20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300, 320, 340, 360, 380],
    'bought': ['yes', 'yes', 'yes', 'no', 'no', 'yes', 'no', 'yes', 'yes', 'no', 'no', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes', 'no']
}

df_raw = pd.DataFrame(data=data)

df = df_raw.copy()

for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float64')    

df.info()

df.describe(include='category').T

le = LabelEncoder()
df['bought'] = le.fit_transform(df['bought'])

df = pd.get_dummies(data=df, drop_first=True)

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df
```
