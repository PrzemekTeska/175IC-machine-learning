```
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data, _ = make_blobs(n_samples=1000, cluster_std=1.2, centers=4,random_state=53)
df = pd.DataFrame(data, columns=['x1','x2'])
px.scatter(df,'x1','x2')

#DBSCAN
cluster = DBSCAN(eps=0.6,min_samples=5)
cluster.fit(data)
df['cluster'] = cluster.labels_
plot = px.scatter(df,'x1','x2','cluster', template='plotly_dark', color_continuous_midpoint=0)
plot.show()

#KMEANS
kmeans = KMeans(n_clusters=4, random_state=53).fit_predict(df)
plt.scatter(df['x1'], df['x2'], c=kmeans)
plt.title("Wykres KMEANS")
plt.show()

#Niestety nie do końca rozumiem jak porównać klastry, więc laborkę wrzucam bez funkcji porównującej :(
```
