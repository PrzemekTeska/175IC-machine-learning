```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from random import randint
import numpy as np


# Zad 1 - import datasetu IRIS i przygotowanie go do modelu
raw_data = load_iris()
all_data=raw_data.copy()

data = all_data['data']
target = all_data['target']
data = np.array(data[:,:2])
target = np.array(list(map(int,target)))
```

```
#Zad 2 - dane podzielone na testowe i uczące
X_train, X_test,y_train,y_test = train_test_split(data,target)
```

```
#Zad 3 - funkcja, która wyświetli wykres na podstawie podanej liczby C oraz typu kernel
def plot(C, kernel):
    plt.figure(figsize=(4,3))
    svc = SVC(C=C, kernel=kernel)
    svc.fit(X_train, y_train)
    plot_decision_regions(X_train, y_train, svc)

    plt.xlabel('petal length (cm)')
    plt.ylabel('sepal width (cm)')
    plt.text(x=3.5,y=5,s=f'C={C}\nKernel={kernel}')
    plt.title(f"SVC train accuracy: {svc.score(X_test,y_test):.4f}")
    plt.show()
```

```
#Zad 4 - zastosowanie funkcji z wyżej. Kernel typu 'precomputed' nie jest sprawdzany, ponieważ dla niego matryca musi być kwadratowa
kernelTypes = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernelTypes:
  plot(randint(1,9),kernel)
  plot(randint(1,9),kernel)
  plot(randint(1,9),kernel)
```
